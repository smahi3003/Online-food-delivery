## Imports

```python
import pandas as pd
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('C:/Users/SMAHI/Desktop/Online-food-delivery/Data/clean_data.csv') 
```

```python
# Preview
df.head()
```

```python
# Shape
df.shape
```

### a. How many customers have placed online food orders again (Output: Yes/No)?

```python
# Count the number of customers who have placed orders again
online_orders_count = df['Output'].value_counts().reset_index()
online_orders_count.columns = ['Output', 'Count']
```

```python
print("Online Orders Placed Again:")
print(online_orders_count)
```

```python
# Bar chart for online orders count
online_orders_chart = alt.Chart(online_orders_count).mark_bar().encode(
    x=alt.X('Output:N',axis = alt.Axis(labelAngle = 0)),
    y=alt.Y('Count:Q'),
    color='Output:N',
    tooltip=['Output:N', 'Count:Q']
).properties(
    title='Placed Online Orders Again',
    width=300
)
```

```python
online_orders_chart
```

### b. What is the feedback distribution (e.g., Positive/Negative)?

```python
# Feedback distribution (e.g., Positive/Negative)?
feedback_distribution = df['Feedback'].value_counts().reset_index()
feedback_distribution.columns = ['Feedback','Count']
```

```python
print("\nFeedback Distribution:")
print(feedback_distribution)
```

```python
# Bar chart for feedback distribution count
feedback_distribution_chart = alt.Chart(feedback_distribution).mark_bar().encode(
    x=alt.X('Feedback:N',axis = alt.Axis(labelAngle = 0)),
    y=alt.Y('Count:Q'),
    color='Feedback:N',
    tooltip=['Feedback:N', 'Count:Q']
).properties(
    title='Feedback Count',
    width=300
)
```

```python
feedback_distribution_chart
```

### c. People who ordered food again and gave positive feedback

```python
# Count the rows where 'Output' is 'Yes' and 'Feedback' is 'Positive'
count_yes_positive = df[(df['Output'] == 'Yes') & (df['Feedback'] == 'Positive')].shape[0]

print(f"People who ordered food again and were satisfied: {count_yes_positive}")
```

```python
print(f"{round((count_yes_positive/df['Output'].value_counts()[0])*100,2)} percent people left positive feedback out of those who ordered again.")
```

### c. Are there any correlations between customer demographics?

```python
# Correlation heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

- As all the numbers are ranging between `-0.2 to 0.2` none of the variables are correlated.

### e. Are there any patterns in feedback based on factors like age, gender, occupation, etc.?

```python
age_feedback_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Age:O',axis = alt.Axis(labelAngle = 0)),
    y='count():Q',
    color=alt.Color('Feedback:N',scale=alt.Scale(range=['red', 'green'])),
    tooltip=['Age:O', 'count():Q', 'Feedback:N']
).properties(
    title='Feedback Distribution by Age',
    width= 400,
    height= 400
)
```

```python
age_feedback_chart
```

### Plot Insight
- People aged 29 & 30 have left negative feedback.
- Highest number of Negative feedback was given by people aged 26.

```python
gender_feedback_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Gender:N',axis = alt.Axis(labelAngle = 0)),
    y='count():Q',
    color=alt.Color('Feedback:N',scale=alt.Scale(range=['red', 'green'])),
    tooltip=['Gender:N', 'count():Q', 'Feedback:N']
).properties(
    title='Feedback Distribution by Gender',
    width= 400,
    height= 400
)
```

```python
gender_feedback_chart
```

### Plot Insights
- More number of negative feedbacks were provided by Males.
- `16.26%` of Females and `19.81%` of Males left Negative feedback.

```python
occupation_feedback_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Occupation:N',axis = alt.Axis(labelAngle = 0)),
    y='count():Q',
    color=alt.Color('Feedback:N',scale=alt.Scale(range=['red', 'green'])),
    tooltip=['Occupation:N', 'count():Q', 'Feedback:N']
).properties(
    title='Feedback Distribution by Occupation',
    width= 400,
    height= 400
)
```

```python
occupation_feedback_chart
```

### Plot Insights
- `33` is the highest count of Negative feedback and it was by Employees.
- Rate of Negative feedback is more in Self Employeed.
- `29.62%` of Self Employeed were not happy with thier food, followed by `27.96%` of Employees and `10.14%` of Students.
